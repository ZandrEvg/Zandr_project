## Анализ торгового ассортимента ##

### ABC - анализ
* определить процент от выручки, который приносит каждая позиция;
* распределить все товары по убыванию этого процента;
* рассчитать по каждой строке долю выручки нарастающим итогом, начиная с тех товаров, которые дают самый существенный вклад;
* разбить все товары на три группы: А - 80% выручки, B - 15% выручки, С - 5% выручки.

```python
# сводная таблица по наименованию
df_product = df.pivot_table(index='product', values=['revenue', 'quantity'], aggfunc='sum')
# расчет доли в процентах от общей выручки
df_product['percent']=df_product['revenue']/df['revenue'].sum()*100
# сортировка по возрастанию
df_product=df_product.sort_values(by='percent', ascending=False)
# доля выручки нарастающим итогом
df_product['cumsum_pers'] = df_product['percent'].cumsum()
df_product_A = df_product.query('cumsum_pers < 80')
df_product_B = df_product.query('80 <= cumsum_pers < 95')
df_product_C = df_product.query('cumsum_pers >= 95')
df_product_A
```

```python
display('Доля группы А от общего колмчества {} процентов'.format(df_product_A.shape[0] / df_product.shape[0] * 100))
display('Доля группы B от общего колмчества {} процентов'.format(df_product_B.shape[0] / df_product.shape[0] * 100))
display('Доля группы C от общего колмчества {} процентов'.format(df_product_C.shape[0] / df_product.shape[0] * 100))
```

Практическое подтверждение закона Поррета, 20% товаров формируют 80% выручки

```python
# внесем данные о категории в основную таблицу
for a in list(df_product.index):
    if a in list(df_product_A.index):
        df_product.loc[a, 'group_ABC'] = 'A'
    elif a in list(df_product_B.index):
        df_product.loc[a, 'group_ABC'] = 'B'
    else:
        df_product.loc[a, 'group_ABC'] = 'C'
    
```

### XYZ - анализ ###
- рассчитать стандартное отклонение
- рассчитать коэффициент вариации (КВ)
- разбить все товары на группы

Для проведения XYZ - выбираем период оценки. Обычно на предприятиях берут период выше периода поставки. В нашем исследовании будем оценивать месячные показатели, т.е расчитывать `коэффициент вариации` по месячным показателям. 

```python
# работа цикла может занимать время более 2 мин
for a in list(df_product.index):
    # очищаем для каждой итерации переменную
    revenue_mon=[]
    # цикл расчета выручки по месяцам
    for mon in range(1,13):
        revenue_mon.append(df.loc[(df['product']==a) & (df['date'].dt.month == mon)]['revenue'].sum())
    # избавляемся от 0. В нашем случае это оправданно, так как мы исследуем влияние спроса, а не качество работы
    # отдела закупа. 0 - не потому что не покупали, а из-за того что его нет в наличии
    revenue_mon = [e for e in revenue_mon if e !=0]
    df_product.loc[a,'ratio_var'] = np.std(revenue_mon) / np.mean(revenue_mon)
df_product
```

```python
# внесем данные о категории в основную таблицу

def group_xyz(_):
    if 0 < _ < 0.1: return('X')
    elif 0.1 <= _ <=0.25:return('Y')
    else:return('Z')

for a in df['product'].unique():
    b=df_product.loc[a,'ratio_var']
    df_product.loc[a, 'group_XYZ'] = group_xyz(b)
```

### Объединенный ABC-XYZ-анализ

Итак, мы разбили товары на три группы по выручке и на три по стабильности спроса. Теперь объединим результаты. Каждый из товаров попадет в одну из 9 групп.

Самый высокий ряд — это категория A, самый нижний — категория C. Чем правее находится товар, тем нестабильнее на него спрос. Левый ряд — это категория X, правый — Z. То есть самые непопулярные товары будут в нижнем правом квадрате, а самые популярные — в верхнем левом.

- Товары из групп `AX` и `BX` обеспечивают основной доход и при этом стабильно продаются. Они всегда должны быть на складе. Необходимый запас легко спрогнозировать, так как объемы продаж почти не меняются от месяца к месяцу. В нашем примере это двуспальная кровать, кухонный стол и диван-книжка.
- Группы `АY` и `BY` — это тоже высокий доход, но стабильность продаж уже ниже. По этим товарам на складе необходимо создать дополнительные резервы, на случай роста спроса в отдельные периоды. Резерв нужен обязательно, это важные для компании группы, которые приносят значительную часть дохода. В примере это шкаф для одежды, шкаф-купе и письменный стол.
- Группы `AZ` и `BZ` тоже приносят существенный доход. Но спрос на них почти невозможно спрогнозировать. Если создавать запасы на складе, исходя из максимально возможной месячной выручки, то значительная часть товара может остаться не проданной. Поэтому лучше использовать другие способы, которые обеспечат нужное количество товара. Например, заключить договоры с поставщиками, которые находятся максимально близко от магазина и могут в любой момент привезти нужную партию. В примере такие товары — это тахта и набор стульев.
- Группа `CX` продается стабильно, но в небольших объемах. По этой позиции нужно создать постоянный запас, исходя из среднего объема продаж.
- Группа `CY` — небольшие объемы продаж и невысокая стабильность. Запасы этой группы нужно создавать по остаточному принципу, т. е. после того, как выделены деньги на закупки товаров из более выгодных групп. Даже если в какой-то момент товара из этой группы не окажется на складе, потери компании будут невелики. В примере это полка.
- Группа `CZ` — самые низкие объемы выручки, при этом спрос невозможно спрогнозировать. Сюда относятся товары, которые уже перестали пользоваться спросом. Запасы по этой категории обычно не создают, работают под заказ. В нашем примере это тумбочка. Но! В эту группу могут попасть и новинки. Они не пользуются спросом потому, что клиенты пока о них не знают. Эти товары еще не привезли в магазины, не выставили на витринах, не запустили маркетинговую кампанию и пр.

Расчитаем обьем запаса товаров для самых востребованных категории и категорий которые возможно прогнозировать(AX, BX, AY, BY, CX)

```python
for _ in list(df_product.index):
    group = df_product.loc[_, 'group_ABC'] + df_product.loc[_, 'group_XYZ']
    if group in ['AX', 'BX', 'AY', 'BY', 'CX']:
        df_product.loc[_, 'month_quantity'] = math.ceil(df_product.loc[_, 'quantity'] / 12)
```

## Формулирование и проверка статистических гипотез

При проведении ABC-анализа, может возникнуть ощущение что более дорогие товары окажутся в категории А, или чем чаше покупают товар, тем с большей вероятностью он окажется в категории А. Отсюда сформулируем две гепотизы
- дорогие товары товары попадают в группу А;
- товары высокого спроса попадают в группу А.

### Проверка гипотезы о цене

Сформулируем гипотезы:
- H0: цена в группах А не имеет статистического отличия от цены в выборке
- H1: цена отличается в большую или меньшую сторону от цен в общей выборке

```python
# сформируем массивы для сравнения
sample1 = df.query('product in list(@df_product_A.index)')['price']
sample2 = df.query('product not in list(@df_product_A.index)')['price']
```

```python
alpha = .05 # критический уровень статистической значимости, если pvalue окажется меньше него, отвергнем гипотезу.

results = st.ttest_ind(sample1, sample2, equal_var = False)
    
print('p-значение: ', results.pvalue)
    
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Проверка гипотезы о количестве

Сформулируем гипотезы
- H0 в группе А товаров заказывают столько же как и в других группах;
- H1 в группе А товаров заказывают больше или меньше чем в других группах.

```python
sample1 = df.query('product in list(@df_product_A.index)')['quantity']
sample2 = df.query('product not in list(@df_product_A.index)')['quantity']
```

```python
alpha = .05 # критический уровень статистической значимости, если pvalue окажется меньше него, отвергнем гипотезу.

results = st.ttest_ind(sample1, sample2, equal_var = False)
    
print('p-значение: ', results.pvalue)
    
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Выводы:** Статистический тест показал, что действительно в группе А присутсвуют товары которые отличаются по цене и интуитивно понятно, что в большую сторноу, также в группу А попадают товары пользующиеся повышенным спросом.
